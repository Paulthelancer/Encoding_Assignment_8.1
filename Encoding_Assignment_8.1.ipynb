```python
df2=df.copy()
```

```python
# Calling libraries to running loops on dataframe column
from pandas.core.dtypes.common import is_numeric_dtype

# Taking a dummy dataframe to temporarily store the One Hot Encoded value for a column
dmy=pd.DataFrame()

# Loooping through each column to check if it is numeric or not
for col in df2.columns:
    if is_numeric_dtype(df2[col]):
        continue
    else:
        #Storing temporary Encoded value to Dummy data frame
        dmy = pd.get_dummies(df2[col],col, drop_first=True)
    
    #droping inplace the associated column that is encoded and stored in dmy
    df2=df2.drop([col],axis=1)
    
    #Concatinating inplace the Encoded colum with the data frame  
    df2=pd.concat([df2,dmy],axis=1)
```

# Label Encoder

```python
df3=df.copy()
df3
# Sex, Region, and Somker are nominal catagorical data. They can only be labeled and can not be orderd.
# Hence we must apply label encoding in these columns.
```

**Hard Coding for Each Column**

```python
from sklearn.preprocessing import LabelEncoder
lab = LabelEncoder()
df3['sex']=lab.fit_transform(df3['sex'])#Label encoding and storing in place
df3['smoker']=lab.fit_transform(df3['smoker'])
df3['region']=lab.fit_transform(df3['region'])
df3
```

**Using For Loop**

```python
df4=df.copy()
from pandas.core.dtypes.common import is_numeric_dtype
from sklearn.preprocessing import LabelEncoder
lab_e = LabelEncoder()

for col in df4.columns:
    if is_numeric_dtype(df4[col]):
        continue
    else:
        df4[col]=lab_e.fit_transform(df4[col])
df4.head(5)
```

**Manual Encoding**

```python
df5=df.copy()
df5['sex'] = df5['sex'].replace(['female', 'male'], [0,1])
df5['smoker'] = df5['smoker'].replace(['yes', 'no'], [1,0])
df5['region'] = df5['region'].replace(['northeast', 'northwest','southeast','southwest'], [0,1,2,3])
df5.head()
```
